```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import lognorm
import statistics as s
```

```python
stat = np.loadtxt('lab3.txt')
```

```python
Perm = stat[:, 4:5]
```

```python
count, bins_count = np.histogram(Perm, bins=20)
pdf = count / sum(count)
a = np.array([0])
pdf = np.append(a, pdf)
bins_count = np.append(a, bins_count)
cdf = np.cumsum(pdf)
plt.plot(bins_count[1:], cdf, label="CDF")
plt.xlabel('permeability')
plt.ylabel('cdf')
plt.savefig('perm_cdf' + '.png')
mean = np.mean(Perm)
print('mean = ' , mean)
variance = np.var(Perm)
print('variance = ' , variance)
median = np.median(Perm)
print('meadian = ' , median)
quartile = s.quantiles(Perm, n=4) 
print('quartiles = ', quartile)
```

```python
plt.plot(bins_count[1:], pdf, label='pdf')
plt.xlabel('permeability')
plt.ylabel('pdf')
plt.savefig('perm_pdf' + '.png')
```
